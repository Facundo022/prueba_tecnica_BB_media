```python
start_time = time.time()
driver = webdriver.Chrome()
driver.get("https://pluto.tv/latam/on-demand")
time.sleep(3)
Pelis_Button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div/div/div/main/div[3]/section/div[2]/ul/li[2]/div/button')))
driver.execute_script("document.body.style.zoom='10%'") 
time.sleep(2)
Pelis_Button.click()
time.sleep(2)

indice = 5
i = 0
data_extracted = []
limite_categoria = ""

try:
    while (limite_categoria != "Mujeres Protagonistas"):
        if i == 0 or i == 4:
            time.sleep(3)
            selector_css = f"section:nth-child({indice}) > span > div > div.category-0-2-253.childCategory > span > a[type='button'][href]"
            # Espera hasta que el elemento sea clickeable usando el selector dinámico
            element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, selector_css)))
            element.click()
            
            ################### Extracion de data de las peliculas #####################

            time.sleep(2)
            categoria = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.TAG_NAME, 'h1'))).text
            url = driver.current_url
            Ul_elementos = driver.find_element(By.XPATH , '//*[@id="overlay-container"]/div/div/div[1]/ul')
            Li_elementos = Ul_elementos.find_elements(By.XPATH, '//li')

            for indice, enlace in enumerate(Li_elementos):
                try:
                    time.sleep(2)
                    enlace = driver.find_elements(By.XPATH, '//ul/li/a[@href]')[indice]
                    print("abrir la pelicula")
                    enlace.click()
                    
                    url = driver.current_url
                    time.sleep(2)

                    try:
                        titulo = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/h2').text
                    except NoSuchElementException:
                        try:
                            titulo = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[1]').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            titulo = 'Titulo no disponible'

                    try:
                        sinopsis = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/section/p ').text
                    except NoSuchElementException:
                        try:
                            sinopsis = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[2]/p').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            sinopsis = 'Sinopsis no disponible'

                    try:
                        informacion_adicional = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[1]/ul').text
                    except NoSuchElementException:
                        try:
                            informacion_adicional = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[2]/ul').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            informacion_adicional = 'informacion adicional no disponible'

                    print("realizando extraccion")

                    informacion_adicional = informacion_adicional.replace('\n',', ')

                    data_extracted.append({
                        'titulo': titulo,
                        'categoria': categoria,
                        'url': url,
                        'sinopsis': sinopsis,
                        'informacion adicional': informacion_adicional
                        })

                    cerrar = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/header/button')
                    cerrar.click()
                    print("cerrar la pelicula")

                except:
                    break

            ################### Finalizacion Extracion de de data de las peliculas #####################
    
            cerrar = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div[1]/div/a')
            cerrar.click()
            indice = 4
            i = 1
            limite_categoria = categoria
            time.sleep(1)

        else:
            time.sleep(4)
            print("dentro de else", indice)
            selector_css = f"section:nth-child({indice}) > span > div > div.category-0-2-253.childCategory > span > a[type='button'][href]"
            # Espera hasta que el elemento sea clickeable usando el selector dinámico
            element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, selector_css)))
            # Obtén la URL del enlace para abrirlo en la nueva pestaña
            enlace = element.get_attribute("href")
            # Abre una nueva pestaña
            driver.execute_script("window.open('');")
            # Cambia el foco a la nueva pestaña
            driver.switch_to.window(driver.window_handles[-1])
            # Abre el enlace en la nueva pestaña
            driver.get(enlace)
            time.sleep(5)
            driver.execute_script("document.body.style.zoom='10%'")

            ################### Extracion de data de las peliculas #####################
            
            time.sleep(2)
            categoria = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.TAG_NAME, 'h1'))).text
            url = driver.current_url
            Ul_elementos = driver.find_element(By.XPATH , '//*[@id="overlay-container"]/div/div/div[1]/ul')
            Li_elementos = Ul_elementos.find_elements(By.XPATH, '//li')

            for indice_pelis, enlace in enumerate(Li_elementos):
                try:
                    time.sleep(2)
                    enlace = driver.find_elements(By.XPATH, '//ul/li/a[@href]')[indice_pelis]
                    print("abrir la pelicula")
                    enlace.click()
                    
                    url = driver.current_url
                    time.sleep(2)

                    try:
                        titulo = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/h2').text
                    except NoSuchElementException:
                        try:
                            titulo = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[1]').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            titulo = 'Titulo no disponible'

                    try:
                        sinopsis = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/section/p ').text
                    except NoSuchElementException:
                        try:
                            sinopsis = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[2]/p').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            sinopsis = 'Sinopsis no disponible'

                    try:
                        informacion_adicional = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[1]/ul').text
                    except NoSuchElementException:
                        try:
                            informacion_adicional = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[2]/ul').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            informacion_adicional = 'informacion adicional no disponible'

                    print("realizando extraccion")

                    informacion_adicional = informacion_adicional.replace('\n',', ')

                    data_extracted.append({
                        'titulo': titulo,
                        'categoria': categoria,
                        'url': url,
                        'sinopsis': sinopsis,
                        'informacion adicional': informacion_adicional
                        })

                    cerrar = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/header/button')
                    cerrar.click()
                    print("cerrar la pelicula")

                except:
                    break

            ################### Finalizacion Extracion de de data de las peliculas #####################
            print(indice)
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
        
        i+=1
        indice+=1
        time.sleep(4)
        print("vuelta:", (i)-1,  categoria, indice, url)
        
except Exception as e:
    print(f"Error: {e}. No se pudo ejecutar correctamente.")

finally:
    print("operacion finalizada")
    driver.close()

    filename = 'Peliculas_data_plutoTV.json'

    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data_extracted, f, ensure_ascii=False, indent=4)

    print(f"Los datos han sido guardados en {filename}.")

        # Marcar el tiempo de fin
    end_time = time.time()

    # Calcular el tiempo de ejecución
    execution_time = end_time - start_time
    print(f"El tiempo de ejecución es: {execution_time:.4f} segundos")
```

<h4> Extraccion de data (TV_vivo)</h4>

```python
start_time = time.time()

data_extracted = []

driver = webdriver.Chrome()
driver.get("https://pluto.tv/latam/live-tv/63eb9255c111bc0008fe6ec4?lang=es")
time.sleep(1)
Tv_live_Button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div/div/div/div[1]/div/button')))
time.sleep(1)
Tv_live_Button.click()
time.sleep(2)
driver.execute_script("document.body.style.zoom='50%'") 

################# entrar en la web y ir a guia ##########################################
time.sleep(2)

# Encuentra el contenedor de las categorías
categorias_container = driver.find_element(By.XPATH, '//*[@id="root"]/div/div/div/main/div[2]/div/section/div[2]/ul')
# Encuentra todas las categorías dentro del contenedor
categorias = categorias_container.find_elements(By.XPATH, '//li')

# entrar en cada categoria
time.sleep(2)
try:
    # listado de categorias
    for indice, categoria in enumerate(categorias):
        time.sleep(2)

        enlace_xpath = f"/html/body/div[1]/div/div/div/main/div[2]/div/section/div[2]/ul/li[{indice+1}]/div/button"
        enlace = driver.find_element(By.XPATH, enlace_xpath) 

        categoria_nombre = enlace.find_element(By.XPATH, './span/span').text  # Utiliza '.' para referirse al elemento actual

        enlace.click()
        time.sleep(1)
        ############################ extracion de data ############################
        
        if indice == 0:
            indice_canales = 2  
        else:
            indice_canales = 6

        while True:
            try:
                print("abriendo canal")
                selector_css = f"#root > div > div > div > main > div.liveTVLayoutContainer-0-2-229.custom-scroll > div > div.channelListContainer-0-2-234 > section > div.list > div > div.channelList-0-2-253.custom-scroll > div > div:nth-child({indice_canales}) > span > div > a[type='button'][href]"
                # Espera hasta que el elemento sea clickeable usando el selector dinámico
                element = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.CSS_SELECTOR, selector_css)))
                element.click() 

                time.sleep(1)

                # Extraer el título
                titulo = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/h2'))
                ).text

                # Extraer la sinopsis
                sinopsis = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[2]/p').text
                
                data_extracted.append({
                    'titulo': titulo,
                    'sinopsis': sinopsis,
                    'categoria': categoria_nombre
                    })

                time.sleep(1)
                print("Realizando extraccion")


                cerrar = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/header/button')
                cerrar.click()
                indice_canales += 1
                time.sleep(1)
                print("cerrando canal")
            except:
                print("Saliendo del bucle.")
                break

            ############################ extracion de data ############################

except Exception as e:
    print(f"Error: {e}. No se pudo ejecutar correctamente.")

finally:
    print("operacion finalizada")
    driver.close()

    filename = 'TV_vivo_data_plutoTV.json'

    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data_extracted, f, ensure_ascii=False, indent=4)

    print(f"Los datos han sido guardados en {filename}.")

    # Marcar el tiempo de fin
    end_time = time.time()

    # Calcular el tiempo de ejecución
    execution_time = end_time - start_time
    print(f"El tiempo de ejecución es: {execution_time:.4f} segundos")
```
