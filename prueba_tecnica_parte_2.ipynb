```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from time import sleep
import json
```

```python
driver = webdriver.Chrome()
driver.get("https://pluto.tv/latam/on-demand")
sleep(3)
Pelis_Button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div/div/div/main/div[3]/section/div[2]/ul/li[5]/div/button')))
driver.execute_script("document.body.style.zoom='10%'") 
sleep(2)
Pelis_Button.click()
sleep(2)

indice = 5
i = 0
data_extracted = []
limite_categoria = ""

try:
    while (limite_categoria != "Series para Maratonear"):
        if i == 0 or i == 4:
            sleep(3)
            selector_css = f"section:nth-child({indice}) > span > div > div.category-0-2-253.childCategory > span > a[type='button'][href]"
            # Espera hasta que el elemento sea clickeable usando el selector dinámico
            element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, selector_css)))
            element.click()
            
            ################### Extracion de data de las peliculas #####################

            sleep(2)
            categoria = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.TAG_NAME, 'h1'))).text
            url = driver.current_url
            Ul_elementos = driver.find_element(By.XPATH , '//*[@id="overlay-container"]/div/div/div[1]/ul')
            Li_elementos = Ul_elementos.find_elements(By.XPATH, '//li')

            for indice, enlace in enumerate(Li_elementos):
                try:
                    sleep(2)
                    enlace = driver.find_elements(By.XPATH, '//ul/li/a[@href]')[indice]
                    print("abrir la pelicula")
                    enlace.click()
                    
                    url = driver.current_url
                    sleep(2)

                    try:
                        titulo = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/h2').text
                    except NoSuchElementException:
                        try:
                            titulo = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[1]').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            titulo = 'Titulo no disponible'

                    try:
                        sinopsis = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/section/p ').text
                    except NoSuchElementException:
                        try:
                            sinopsis = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[2]/p').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            sinopsis = 'Sinopsis no disponible'

                    try:
                        informacion_adicional = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[1]/ul').text
                    except NoSuchElementException:
                        try:
                            informacion_adicional = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[2]/ul').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            informacion_adicional = 'informacion adicional no disponible'

                    print("realizando extraccion")

                    informacion_adicional = informacion_adicional.replace('\n',', ')

                    data_extracted.append({
                        'titulo': titulo,
                        'categoria': categoria,
                        'url': url,
                        'sinopsis': sinopsis,
                        'informacion adicional': informacion_adicional
                        })

                    cerrar = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/header/button')
                    cerrar.click()
                    print("cerrar la pelicula")

                except:
                    break

            ################### Finalizacion Extracion de de data de las peliculas #####################
    
            cerrar = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div[1]/div/a')
            cerrar.click()
            indice = 4
            i = 1
            limite_categoria = categoria
            sleep(1)

        else:
            sleep(4)
            print("dentro de else", indice)
            selector_css = f"section:nth-child({indice}) > span > div > div.category-0-2-253.childCategory > span > a[type='button'][href]"
            # Espera hasta que el elemento sea clickeable usando el selector dinámico
            element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, selector_css)))
            # Obtén la URL del enlace para abrirlo en la nueva pestaña
            enlace = element.get_attribute("href")
            # Abre una nueva pestaña
            driver.execute_script("window.open('');")
            # Cambia el foco a la nueva pestaña
            driver.switch_to.window(driver.window_handles[-1])
            # Abre el enlace en la nueva pestaña
            driver.get(enlace)
            sleep(5)
            driver.execute_script("document.body.style.zoom='10%'")

            ################### Extracion de data de las peliculas #####################
            
            sleep(2)
            categoria = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.TAG_NAME, 'h1'))).text
            url = driver.current_url
            Ul_elementos = driver.find_element(By.XPATH , '//*[@id="overlay-container"]/div/div/div[1]/ul')
            Li_elementos = Ul_elementos.find_elements(By.XPATH, '//li')

            for indice_pelis, enlace in enumerate(Li_elementos):
                try:
                    sleep(2)
                    enlace = driver.find_elements(By.XPATH, '//ul/li/a[@href]')[indice_pelis]
                    print("abrir la pelicula")
                    enlace.click()
                    
                    url = driver.current_url
                    sleep(2)

                    try:
                        titulo = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/h2').text
                    except NoSuchElementException:
                        try:
                            titulo = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[1]').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            titulo = 'Titulo no disponible'

                    try:
                        sinopsis = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/section/p ').text
                    except NoSuchElementException:
                        try:
                            sinopsis = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[2]/p').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            sinopsis = 'Sinopsis no disponible'

                    try:
                        informacion_adicional = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[1]/ul').text
                    except NoSuchElementException:
                        try:
                            informacion_adicional = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/section/div/div[1]/div[2]/ul').text
                        except NoSuchElementException:
                            # Si tampoco se encuentra en la segunda ubicación, asignar un valor por defecto
                            informacion_adicional = 'informacion adicional no disponible'

                    print("realizando extraccion")

                    informacion_adicional = informacion_adicional.replace('\n',', ')

                    data_extracted.append({
                        'titulo': titulo,
                        'categoria': categoria,
                        'url': url,
                        'sinopsis': sinopsis,
                        'informacion adicional': informacion_adicional
                        })

                    cerrar = driver.find_element(By.XPATH, '//*[@id="overlay-container"]/div/div/div/header/button')
                    cerrar.click()
                    print("cerrar la pelicula")

                except:
                    break

            ################### Finalizacion Extracion de de data de las peliculas #####################
            print(indice)
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
        
        i+=1
        indice+=1
        sleep(4)
        print("vuelta:", (i)-1,  categoria, indice, url)
        
except Exception as e:
    print(f"Error: {e}. No se pudo ejecutar correctamente.")

finally:
    print("operacion finalizada")
    driver.close()

    filename = 'Series_data_plutoTV.json'

    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data_extracted, f, ensure_ascii=False, indent=4)

    print(f"Los datos han sido guardados en {filename}.")
```
